#Relatório de Vendas

```
import pandas as pd
```

## Importando Base de Dados

```
planilha_vendas = pd.read_excel('Vendas.xlsx')
planilha_vendas.head(5)
```

##Faturamento por Loja

```
planilha_faturamento = planilha_vendas[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()
planilha_faturamento = planilha_faturamento.sort_values(by = 'Valor Final', ascending = False)
planilha_faturamento.head()
```

##Quantidade Vendida por Loja

```
planilha_quantidade = planilha_vendas[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()
planilha_quantidade = planilha_quantidade.sort_values(by = 'Quantidade', ascending = False)
planilha_quantidade.head()
```

##Ticket Médio

```
ticket_medio = (planilha_faturamento['Valor Final'] / planilha_quantidade['Quantidade']).to_frame().round(2)
ticket_medio = ticket_medio.rename(columns = {0 : 'Ticket Medio'})
ticket_medio = ticket_medio.sort_values(by = 'Ticket Medio', ascending = False)
ticket_medio
```

##Enviando E-mail

```
import smtplib
import email.message

def funcao_email(resumo_loja, loja):

  server = smtplib.SMTP('smtp.gmail.com:587')  
  corpo_email = f"""<p>Prezados,</p>
  {resumo_loja.to_html()}
  <p>Grato desde já,</p>
  <p></p>"""
    
  msg = email.message.Message()
  msg['Subject'] = f"Registo Financeiro - Loja {loja}"
    
  # Fazer antes (apenas na 1ª vez): Ativar Aplicativos não Seguros.
  # Gerenciar Conta Google -> Segurança -> Aplicativos não Seguros -> Habilitar
  # Caso mesmo assim dê o erro: smtplib.SMTPAuthenticationError: (534,
  # Você faz o login no seu e-mail e depois entra em: https://accounts.google.com/DisplayUnlockCaptcha
  msg['From'] = ''
  msg['To'] = ''
  password = ****
  msg.add_header('Content-Type', 'text/html')
  msg.set_payload(corpo_email)
    
  s = smtplib.SMTP('smtp.gmail.com: 587')
  s.starttls()
  # Login Credentials for sending the mail
  s.login(msg['From'], password)
  s.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
print('Email enviado')
```

##Criando Relatório Individual por Loja

```
lojas = planilha_vendas['ID Loja'].unique()

for loja in lojas:
  planilha_loja = planilha_vendas.loc[planilha_vendas['ID Loja'] == loja, ['ID Loja', 'Quantidade', 'Valor Final']]
  resumo_loja = planilha_loja.groupby('ID Loja').sum()
  resumo_loja['Ticket Médio'] = resumo_loja['Valor Final'] / resumo_loja['Quantidade']
  funcao_email(resumo_loja, loja)
```

##E-mail para Diretoria

```
planilha_diretoria = planilha_faturamento.join(planilha_quantidade).join(ticket_medio)
funcao_email(planilha_diretoria, "Todas as Lojas")
```
